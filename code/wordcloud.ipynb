```python
# import required packages/dependencies
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

**Load Data**

cleaned_ufo_au.csv

```python
# Import the csv file 
comments = pd.read_csv("Resources/cleaned_ufo_au.csv")
comments.head()
```

```python
text = ' '.join(comments['comments'].astype(str).tolist())
```

```python
wordcloud = WordCloud().generate(text)
```

```python
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.figure(dpi=300)
plt.show()
```

```python
wordcloud.to_file("wordcloud.png")
```
